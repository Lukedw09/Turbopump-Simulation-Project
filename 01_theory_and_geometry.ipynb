## Theory Background
### Euler Pump Equation
The Euler pump equation is derived from the conservation of angular momentum and follows the form:

$Y = C_{u2}U_2 - C_{u1}U_1$

Where $Y$ represents the energy required to pump the fluid per unit mass, and all other variables are as defined above. Euler created his famous pump equation in 1756 whilst exploring the field of fluid dynamics. This equation will be fundamental in calculating the maximium energy that we can achieve with given inlet/outlet speeds for our centrifugal pump.
### Ideal Case for Maximum Head
Under the assumptions active in this notebook, $C_{u1} = 0$ and Euler's pump equation can be simplified to

$Y_{Ideal} = C_{u2}U_2$

We can divide both sides of this equation by $g_0$, the gravitational constant, to achieve our maximum possible head, $H_{Ideal}$.

$H_{Ideal}$ = $\frac{1}{g_0}(C_{u2}U_2)$

This will be the equation we use to calculate the upper limit for our refrence and model pumps.

## Calculations
Our values are calculated in the code below.

```python
import numpy as np
#Implement our variables from the Reference Pump
r_2 = 0.09445 #meters
N = 3000 #rpm - convert
g0 = 9.80 #m*s^-2

#Convert rpm to rad/s.
N_conv = N*((2*np.pi)/60)

#Note that U2 = Uc2, which can be determnined from the angular acceleration and the impeller radius at exit.
U2 = N_conv*r_2
Uc2 = U2

#Applying Euler's pump equation for no pre-rotation.
H_ideal = (1/g0)*(U2*Uc2)

#Printing our output
print(f"The ideal head is {H_ideal} meters")
```
